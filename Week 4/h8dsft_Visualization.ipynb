**CURRENCY :**

```python
fifa['Value'] = fifa['Value'].str.replace('€', '')
fifa['Wage'] = fifa['Wage'].str.replace('€', '')
fifa['Release Clause'] = fifa['Release Clause'].str.replace('€', '')
```

```python
print(f"Value: \n{fifa['Value'].unique().tolist()}")
print(f"Wage: \n{fifa['Wage'].unique().tolist()}")
print(f"Release Clause: \n{fifa['Release Clause'].unique().tolist()}")
```

```python
def convert_currency(x):
    nominal = 0
    if type(x) == str:
        if str(x[-1]) == 'K':
            nominal = float(x[:-1]) * 1000
        elif str(x[-1]) == 'M':
            nominal = float(x[:-1]) * 1000000
        else:
            nominal = float(x)
    else:
        nominal = np.nan
    return nominal
```

```python
fifa['Value'] = fifa.apply(lambda x: convert_currency(x['Value']), axis = 1)
fifa['Wage'] = fifa.apply(lambda x: convert_currency(x['Wage']), axis = 1)
fifa['Release Clause'] = fifa.apply(lambda x: convert_currency(x['Release Clause']), axis = 1)
```

**HEIGHT & WEIGHT :**

```python
fifa['Height'].unique()
```

```python
def height_to_cm(x):
    if type(x) == str:
        feet_to_inch = int(x[0]) * 12
        inch = int(x[2:])
        height = (feet_to_inch + inch) * 2.54
    else:
        height = np.nan
    return height
```

```python
fifa['Height'] = fifa.apply(lambda x: height_to_cm(x['Height']), axis = 1)
```

```python
fifa['Height'].unique()
```

```python
def weight_to_number(x):
    if type(x) == str:
        weight = int(x[:-3])
    else:
        weight = np.nan
    return weight
```

```python
fifa['Weight'] = fifa.apply(lambda x: weight_to_number(x['Weight']), axis = 1)
```

```python
fifa['Weight'].unique()
```

##**BAR PLOT :**

```python
portugal_position = fifa[fifa['Nationality'] == 'Portugal'].groupby(['Position']).count()['ID'].reset_index().set_index('Position').sort_values('ID', ascending = False)
portugal_position.columns = ['count']
```

```python
portugal_position.plot.bar(title = 'Position for Most of Portuguese', xlabel = 'Position', ylabel= 'frequency', rot=0, figsize = (15,7))
plt.show()
```

```python
plt.figure(figsize = (15,7))
plt.bar(data = portugal_position, height='count', x = portugal_position.index)
plt.title('Position for Most of Portuguese')
plt.xlabel('Position')
plt.ylabel('frequency')
plt.show()
```

```python
club_median_wage = fifa.groupby(['Club']).median()[['Wage']].sort_values('Wage', ascending = False)[:10].sort_values('Wage', ascending = True)
club_median_wage.columns = ['Median Wage']
```

```python
club_median_wage.plot.barh(title = 'Top 10 Club by Median Wage', xlabel = '', ylabel= 'Median Wage', rot=0, figsize = (15,8))
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.barh(data = club_median_wage, width='Median Wage', y = club_median_wage.index)
plt.title('Top 10 Club by Median Wage')
plt.xlabel('Median Wage')
plt.ylabel('')
plt.show()
```

##**MAPS FOLIUM :**

```python
world_map = folium.Map()
world_map
```

```python
map = fifa[['Nationality', 'Wage']]
```

```python
map.head(10)
```

```python
map['Nationality'].unique()
```

```python
#Setting
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
world_geo = f'{url}/world-countries.json'
```

```python
folium.Choropleth(
    geo_data = world_geo,
    data = map,
    columns = ['Nationality', 'Wage'],
    key_on = 'feature.properties.name',
    fill_color = 'Blues_r',
    fill_opacity = 0.7,
    line_opacity = 0.2,
    legend_name = "Player Wage by Country"
).add_to(world_map)
world_map
```
