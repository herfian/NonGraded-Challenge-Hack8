##**BAR PLOT :**

Membuat bar plot dengan library seaborn, kolom yang digunakan adalah kolom Position.

```python
plt.figure(figsize=(15, 7))
sns.countplot(x = 'Position', data = fc, order = fc['Position'].value_counts().index)
plt.show()
```

Kemduian membuat plot dengan menggunakan 2 sumbu, sumbu x untuk kolom position dan sumbu y untuk kolom wage.

```python
plt.figure(figsize=(15, 7))
sns.barplot(x = 'Position', y = 'Wage', data = fc, estimator = np.median)
plt.show()
```

##**HISTOGRAM, KDE, DENSITY :**

Membuat Histogram, untuk melihat usia rentang usia mana yang lebih dominan, pada plot ini, rentang usia player fifa yang paling dominan adalah usia 20 sampai 30.

```python
plt.figure(figsize=(15, 7))
sns.histplot(x = fc['Age'], bins = 10)
plt.show()
```

Menambahkan garis pada Histogram.

```python
plt.figure(figsize=(15, 7))
sns.histplot(x = fc['Age'], bins = 10, kde = True)
plt.show()
```

Plot tanpa Histogram dan hanya garis.

```python
plt.figure(figsize=(15, 7))
sns.kdeplot(x = fc['Age'])
plt.show()
```

Contoh lain Histogram Bar, menggunakan 2 element kolom yaitu Height dan Preferred Foot.

```python
plt.figure(figsize=(15, 7))
sns.histplot(x = fc['Height'], hue = fc['Preferred Foot'], bins = 10)
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.histplot(x = fc['Height'], hue = fc['Preferred Foot'], bins = 10, kde = True)
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.kdeplot(x = fc['Height'], hue = fc['Preferred Foot'])
plt.show()
```

##**PIE CHART :**

Membuat Pie chat untuk Preferred Foot yang hanya memiliki 2 nilai yaitu Left and Right. Untuk memvisualisasikan persentasenya, menggunakan command seperti dibawah ini:

```python
fc['Preferred Foot'].value_counts().plot.pie(explode=[0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(10,7))
plt.show()
```

##**PAIR PLOT :**

Seperti namanya yaitu Pair, Tujuan dari Plot ini adalah untuk melihat kecocokan antara beberapa kolom yang memiliki nilai angka atau tipe data number, disni saya menggunakan 6 kolom untuk muntuk dijadikan pair plot.

```python
pair_plot = fc[['Age', 'Value', 'Wage', 'Overall', 'Potential', 'Preferred Foot']]
sns.pairplot(
    pair_plot, hue = 'Preferred Foot'
)
plt.show()
```

##**VIOLIN PLOT :**

Menggunakan Plot Violin untuk melihat angka pesebaran nilai pada kolom Age.

```python
plt.figure(figsize=(15, 7))
sns.violinplot(
    x = fc['Age']
)
plt.show()
```

Contoh lain untuk melihat pesebaran angka pada kolom Potential.

```python
plt.figure(figsize=(15, 7))
sns.violinplot(
    x = fc['Potential']
)
plt.show()
```
