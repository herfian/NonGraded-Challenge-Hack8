##**Solution 2 :**

```python
tambah = A+B #Penjumlahan Vektor A + Vektor B
print(tambah)
```

```python
kurang = B-C #Pengurangan Vektor B + Vektor C
print(kurang)
```

```python
dot1=A.dot(C) # Vektor A dot product Vektor C cara 1
print(dot1)
```

```python
dot2=np.dot(A,C) # Vektor A dot product Vektor C cara 2
print(dot2)
```

```python
dot3 = A@C # Vektor A dot product Vektor C cara 3 (Terbaru)
print(dot3)
```

```python
kali = A*B #Perkalian vektor A dengan vektor B
print(kali)
```

```python
Norm=np.linalg.norm(A,2) #Menghitung vektor A L2
print(Norm)
```

Membuat sudut antara vektor A dan B dengan perintah dibawah ini, pertama Vektor A dot product dengan vektor B kemudian dibagi dengan hasil perkalian antara norm vektor A dan B:

```python
cosine=(A@B)/(np.linalg.norm(A) * np.linalg.norm(B))
print(cosine)
```

Jika sudah tambahkan perintah `arccos` kemudian print.

```python
theta = np.arccos(cosine)
print(theta)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
u = np.array([2, 5])
v = np.array([3, 1])
```

```python
x_coords, y_coords = zip(u, v)
plt.scatter(x_coords, y_coords, color=["r","b"])
plt.axis([0, 9, 0, 6])
plt.grid()
plt.show()
```

```python
def plot_vector2d(vector2d, origin=[0, 0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],
              head_width=0.2, head_length=0.3, length_includes_head=True,
              **options)
```

```python
plot_vector2d(u, color="r")
plot_vector2d(v, color="b")
plt.axis([0, 9, 0, 6])
plt.grid()
plt.show()
```
