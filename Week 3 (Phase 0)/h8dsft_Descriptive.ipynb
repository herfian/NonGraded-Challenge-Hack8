###**PROBLEM 1:**

Cari Measure of Central Tendency dari data tersebut. Sebagai catatan, yang dicari disini adalah nilai mean, median dan mode pada keseluruhan data, yang perlu digarisbawahi adalah, data yang memiliki tipe data Object hanya perlu dicari modenya saja dan data yang memiliki tipe data number (int,float,etc) dicari ketiganya.

**STEP 1 :**

Mencari Mode pada keseluruhan kolom data yang memiliki tipe data object dengan menggunakan perintah `.mode()`.

```python
print ('Mode of Date                          : ' + str(data_clean['Date'].mode()[0]))
print ('Mode of Location ISO Code             : ' + str(data_clean['Location ISO Code'].mode()[0]))
print ('Mode of Location                      : ' + str(data_clean['Location'].mode()[0]))
print ('Mode of Location Level                : ' + str(data_clean['Location Level'].mode()[0]))
print ('Mode of Province                      : ' + str(data_clean['Province'].mode()[0]))
print ('Mode of Country                       : ' + str(data_clean['Country'].mode()[0]))
print ('Mode of Continent                     : ' + str(data_clean['Continent'].mode()[0]))
print ('Mode of Island                        : ' + str(data_clean['Island'].mode()[0]))
print ('Mode of Time Zone                     : ' + str(data_clean['Island'].mode()[0]))
print ('Mode of Special Status                : ' + str(data_clean['Special Status'].mode()[0]))
print ('Mode of Total Cities                  : ' + str(data_clean['Total Cities'].mode()[0]))
print ('Mode of Total Urban Villages          : ' + str(data_clean['Total Urban Villages'].mode()[0]))
print ('Mode of Total Rural Villages          : ' + str(data_clean['Total Rural Villages'].mode()[0]))
print ('Mode of Case Fatality Rate            : ' + str(data_clean['Case Fatality Rate'].mode()[0]))
print ('Mode of Case Recovered Rate           : ' + str(data_clean['Case Recovered Rate'].mode()[0]))
print ('Mode of Growth Factor of New Cases    : ' + str(data_clean['Growth Factor of New Cases'].mode()[0]))
print ('Mode of Growth Factor of New Deaths   : ' + str(data_clean['Growth Factor of New Deaths'].mode()[0]))
```

Kemudian mencari meannya, cara dibawah ini adalah cara manual sama seperti daiatas.

```python
print ('Mean of New Cases : ' + str(data_clean['New Cases'].mean()))
print ('New Deaths        : ' + str(data_clean['New Deaths'].mean()))
print ('New Recovered     : ' + str(data_clean['New Recovered'].mean()))
```

Berikut command cara cepatnya, perintah `.mean() `untuk mencari mean dan `.median()` untuk mencari mediannya.

```python
print("{} \t{} \t{} \t{}".format("Column Name", "Mean", "Median", "Mode").expandtabs(30))
print("-"*100)
for col in data_clean.select_dtypes("number"):
    print("{} \t{} \t{} \t{}".format(col, data_clean[col].mean(), data_clean[col].median(),data_clean[col].mode()).expandtabs(30))
```
